### Introduction

In this notebook, we will develop an automated method to find the times where a machine will change from a normal operating mode into faulty and failed modes. The data we have available to us is in the form of 20 .csv files. Each file represents a different machine with different fault and failure times, and reports 3000 total samples with each sample containing data from four different sources.

We would like to impose several constraints on our solution based on the nature of our data and the description of the task. Firstly, the data we are given is not labeled with fault and failure times, so our solution should be an unsupervised algorithm, and should not rely on manual labeling of the times.

Secondly, our algorithm should be an *autoregressive* algorithm, which means it should only take into account past and present data when making predictions. In the context of our problem, it does not make any sense to include future data into our calculations, since we are trying to find points of fault and failure as they occur. Predicting when a machine has failed after it has failed does not help the company to fix the machine before fault occurs.

Finally, we want our algorithm to be able to have the ability to perform *online* prediction, which means it should have the ability to process input piece by piece as if the input was being streamed into the model. This aligns with our task description, as we want to be able to predict if a machine has failed while the machine is sending the data.

Below, I have created several transformations that will help to denoise and remove outliers from the data. Again, since we cannot assume normality, we should use non-parametric methods.

```python
class Transform(object):
    """ Represents a transformation applied to an array of samples.
    
    Args:
        data (np.ndarray): An (MxN) sample of floats. M represents samples, and N represents features of each sample.
    Returns:
        The data, unchanged.
    """
    def __init__(self):
        pass
    
    def __call__(self, data : np.ndarray) -> np.ndarray:
        return data

class Compose(Transform):
    """ Sequentially applies multiple transforms to a single sequence of data.
    Args:
        transforms (List): A list of transforms to apply to the data.
        data (np.ndarray): An (MxN) sample of floats. M represents samples, and N represents features of each sample.
    Returns:
        The data, with each transform applied to it in order.
    """
    def __init__(self, transforms=[]):
        super(Compose, self).__init__()
        self.transforms = transforms
        
    def __call__(self, data : np.ndarray) -> np.ndarray:
        for trf in self.transforms:
            data = trf(data)
        return data
    
class RemoveOutliers(Transform):
    """ Removes outlier values based on the number of Mean Absolute Deviations (MAD) from the median of the data.
        The MAD is calculated by taking the median(abs(X - median(X))), and is functionally similar to the
        standard deviation from the mean.
        
    Args:
        mad_deviations (int): The cutoff number of MADs away from the median a value can be to be classified
                              as an outlier.
        data (np.ndarray): An (MxN) sample of floats. M represents samples, and N represents features of each sample.
                          
    Returns:
        An (MxN) np.ndarray representing the same data, but with outliers set to the median of the data.
    """
    def __init__(self, mad_deviations:int=7):
        super(RemoveOutliers, self).__init__()
        self.mad_deviations = mad_deviations
        
    def __call__(self, data : np.ndarray) -> np.ndarray:
        medians = np.median(data, axis=0)
        mad = np.median(np.abs(data - medians), axis=0)

        def replace_outliers(data, idx):
            return np.where(np.abs(data[:,idx] - medians[idx]) > self.mad_deviations * mad[idx],
                            np.nan, data[:,idx])
            
        nan_outliers = np.array([replace_outliers(data, i) for i in range(data.shape[1])]).T
        new_medians = np.nanmedian(nan_outliers, axis=0)

        def replace_nans(data, idx):
            return np.where(np.isnan(data[:,idx]), new_medians[idx], data[:,idx])
    
        return np.array([replace_nans(nan_outliers, i) for i in range(nan_outliers.shape[1])]).T
    
class AbsoluteValue(Transform):
    """ Returns the absolute values of the data.
    
    Args:
        data (np.ndarray): An (MxN) sample of floats. M represents samples, and N represents features of each sample.
    Returns:
        The data, with all negative values flipped.
    """
    def __init__(self):
        super(AbsoluteValue, self).__init__()

    def __call__(self, data : np.ndarray) -> np.ndarray:
        return np.abs(data)
    
class MedianFilter(Transform):
    """ Applies a sliding window filter to the data, taking the median of all elements in the window and
        moving this window across the entirety of the data.
        
    Args:
        window (int): The size of the window. Must be an odd number in order to break ties.
        data (np.ndarray): An (MxN) sample of floats. M represents samples, and N represents features of each sample.
    Returns:
        An (MxN) array of medians. The data is padded from the front and back so that the output is the same shape
        as the input.
    """
    def __init__(self, window : int):
        super(MedianFilter, self).__init__()
        assert(window % 2 == 1)
        self.window = window
        
    def __call__(self, data : np.ndarray) -> np.ndarray:
        data_df = pd.DataFrame(data)
        front_pad = data_df.iloc[0:self.window // 2]
        back_pad = data_df.iloc[len(data_df) - self.window // 2:]
        data_df = pd.concat([front_pad, data_df, back_pad], axis=0)
        
        medians = data_df.rolling(window=self.window, center=True).median()
        return medians.iloc[self.window // 2:len(medians) - self.window // 2].values
    
class MeanFilter(Transform):
    """ Applies a sliding window filter to the data, taking the mean of all elements in the window and
        moving this window across the entirety of the data.
        
    Args:
        window (int): The size of the window. Must be an odd number to be consistent with the MedianFilter.
        data (np.ndarray): An (MxN) sample of floats. M represents samples, and N represents features of each sample.
    Returns:
        An (MxN) array of means. The data is padded from the front and back so that the output is the same shape
        as the input.
    """
    def __init__(self, window : int):
        super(MeanFilter, self).__init__()
        assert(window % 2 == 1)
        self.window = window
        
    def __call__(self, data : np.ndarray) -> np.ndarray:
        data_df = pd.DataFrame(data)
        front_pad = data_df.iloc[0:self.window // 2]
        back_pad = data_df.iloc[len(data_df) - self.window // 2:]
        data_df = pd.concat([front_pad, data_df, back_pad], axis=0)
        
        means = data_df.rolling(window=self.window, center=True).mean()
        return means.iloc[self.window // 2:len(means) - self.window // 2].values

class Discretize(Transform):
    """ Rounds all values in the data down to its integer representation.
    
    Args:
        data (np.ndarray): An (MxN) sample of floats. M represents samples, and N represents features of each sample.
    Returns:
        An (MxN) sample of ints.
    """
    def __init__(self):
        super(Discretize, self).__init__()

    def __call__(self, data : np.ndarray) -> np.ndarray:
        return data.astype(np.int16)
```

### Approach

Now we will develop our approach to detect the times when the machine starts to fault and fail, called "change points". Once again, our approach will need to be non-parametric, which means it works with data that does not fit a known distribution. We also want our approach to have the ability to work with online data, so we need to add the ability to predict one sample at a time.

Simply put, our approach works by taking a small segment of the machine data, and compares each half of the segment. If the two halves are different from each other, then that means that our machine state has changed from normal to faulty or from faulty to failed. If the two halves are similar, then it means that the machine has not changed its state. We divide the entirety of the data into these segments and perform this comparison for every segment in order to find when a change occurs.

Our model will be based on a sliding window algorithm. We will split the data inside the window into halves, and perform a KolmogorovSmirnov test to see if the two halves have equal probability distributions. If they do, then we can extrapolate that no change in machine state has occurred. However, if the two halves have different distributions, then we can assume the machine has changed state from (normal -> faulty) or (faulty -> failed). We choose the Kolmogorov-Smirnov test because it makes no assumptions about data distribution.

```python
class ChangePointDetector(object):
    """ An online approach for detecting fault and failure points using two adjacent sliding windows. If the samples
        represented by each window are found to not be drawn from the same distribution, then we 
        can conclude a change in the machine operating mode has occurred. We determine if two samples
        have the same distribution by performing a Kolmogorov-Smirnov test.
    """
    
    def __init__(self, window_len:int=120, step:int=1, transform:Transform=None, data:np.ndarray=None):
        super(ChangePointDetector, self).__init__()
        self.window_len = window_len
        self.step = step
        self.transform = transform
        if data is not None:
            assert(len(data) == window_len)
        self.data = data
        
    def __call__(self, data : np.ndarray) -> np.ndarray:
        """ Evaluate a long sequence of data by sliding a window across the data.
        
        Args:
            data (np.ndarray): An (MxN) array of samples. Each row is a separate sample, and each column is a
                               different feature.
        Returns:
            A 1d array of probabilities range(0,1) of length M. 
        """
        if self.data is not None:
            data = np.concatenate((self.data, data), axis=0)
        else:
            assert(len(data) >= self.window_len)
            data = np.concatenate((data[0:self.window_len], data), axis=0)
        preds = []
        for idx in range(0, len(data) - self.window_len, self.step):
            preds.append(self._evaluate_window(data[idx:idx + self.window_len]))
        return np.array(preds)
    
    def stream(self, sample : np.ndarray) -> float:
        """ Evaluates the probability that the machine state has changed with the addition of a single
            new data point.
        Args:
            sample (np.ndarray): A single sample measurement. Should have the same shape as samples already stored
                                 in the model. Additionally, the model should already have stored data prior to
                                 prediction.
        Returns:
            the probability that the machine has faulted or failed, based on the addition of this new data point.
        """
        assert(self.data is not None and len(self.data) >= self.window_len)
        
        sample = sample.reshape(1,-1)
        self.data = np.append(self.data, sample, axis=0)
        return self._evaluate_window(self.data[-self.window_len:])
    
    def _evaluate_window(self, data: np.ndarray) -> float:
        """ Compares the two halves of the data and tests if they are drawn from the same distribution.
        Args:
            data (np.ndarray): A sequence of samples. Should be equal in length to the window size.
        Returns:
            The probability that the second half of the data is drawn from a different distribution from the
            first half of the data. If there is more than 1 feature, returns the average of probabilities for
            each feature.
        """
        window_0 = data[0:self.window_len // 2]
        window_1 = data[self.window_len // 2: ]
        if self.transform:
            window_0 = self.transform(window_0)
            window_1 = self.transform(window_1)
        p_sum = 0;
        for feature in range(data.shape[1]):
            stat, p = stats.kstest(window_0[feature], window_1[feature])
            p_sum += p
        return 1 - (p_sum / data.shape[1])
    
    def threshold_probs(self, probs:np.ndarray, transform:Transform=None, p_thresh:float=0.95, scale:int=1):
        """ Transforms an array of probabilities to an array of binary values.
        Args:
            probs (np.ndarray): The array of probabilities.
            transform (Transform): An optional transform to apply to the probabilities.
            p_thresh (float): The threshold to determine if a probability turns into 1 or 0.
            scale (int): Instead of an array of [1,0], can return an array of [scale, 0]
        Returns:
            an nd.nparray with the same shape as the input, with probabilities transformed into one of two values.
        """
        if transform:
            probs = transform(probs)
        return np.where(probs > p_thresh, scale, 0)
```

### Evaluation

Here, we will test our algorithm with the data from several machines:

```python
transforms = Compose([
    RemoveOutliers(mad_deviations=8),
    AbsoluteValue(),
    MedianFilter(window=7),
    Discretize(),
])
detector = ChangePointDetector(window_len=90, step=1, transform=transforms)
```

```python
median = MedianFilter(window=9)
```

```python
for i in range(1, 20, 2):
    machine = pd.read_csv('exampleco_data/machine_' + str(i) + '.csv', index_col=0).values
    plt.title("Machine " + str(i))
    plt.xlabel("Time")
    plt.plot(range(0,3000), median(machine))
    plt.plot(range(0,3000), detector.threshold_probs(detector(machine), 
                                                     transform=None, p_thresh=0.8, scale=150) - 75)
    plt.show()
```

In the above graphs, the machine data has had its outliers removed for readability purposes. The purple vertical lines on the graph denote where the algorithm has detected a change in operating mode has occurred. We can see that most of the graphs have two lines - One near the center of the faulty mode, and then one directly after the faulty mode. We are able to detect the failure mode for all machines accurately, but sometimes misidentify the fault mode. 

We can change around the hyperparameters of the model to improve our performance and better identify the fault mode.

- Firstly, we can choose to include some prior data into our model, perhaps from a different machine that we know is functioning normally. This will allow us to identify the fault mode in the situation that the machine starts out in fault mode.

- The window size affects the number of samples that we consider when performing our statistical tests. Increasing the window size will decrease the number of false negatives (classifying a point in time as not a change point when it actually is) but will also result in a more delayed prediction, where there is a larger time lag between when the fault mode actually starts and when the model predicts it starts. Generally, I've found that a window size that is a bit smaller than the estimated length of the machine's fault mode is ideal.

- The probability threshold will change the value a probability needs to be above to be considered a change point. If we decrease the threshold, we will get more total positive predictions.

- Finally, we can also change the parameters of our transforms to increase or decrease the amount of noise we remove from the data. The more noise we remove, the better our model will be able to generalize. However, if we remove too much noise then our model won't be able to make accurate predictions. This is an instance of the bias vs. variance tradeoff.

Below, we see an example of streaming data to the model:

```python
detector = ChangePointDetector(window_len=90, step=1, transform=transforms, data=machine1.iloc[0:90].values)

for t in range(0,3000):
    p = detector.stream(machine1.iloc[t].values)
    if p > 0.8:
        print(f"Time: {t} is a change point with probability {p}")
```

### Discussion

One strength of this approach is its interpretability. Since a sliding window is a simple approach, it is easy to explain how the model works to a non-technical audience. Another strength is that it requires no training on labelled data. This means that the model can be immediately deployed for any machine with just some tweaking of the hyperparameters. This model also makes no assumptions about the distribution of the data, which means that it can generalize well to any machine. Finally, this method doesn't "cheat" by leaking future data into the calculation.

A limitation of this method is that it works with an unlabeled time series, and so it cannot identify *which* mode a machine is running in. It can only detect changes in modes. Another limitation is that it is unable to identify a mode change as soon as it occurs. Instead, the predicted time will lag behind the actual time, and this is dependent on the window size chosen.

If I continued with this exercise, I would have liked to try implementing a Bayesian approach like the approach in this paper: https://arxiv.org/pdf/0710.3742.pdf
The approach described in the paper would not have the "time lag" limitation, so there would be no delay when predicting the change point. It also is an online algorithm, which is necessary as per the problem description. However, it is a parametric approach, and so if our data was composed of multiple different distributions it would not perform optimally. Finally, the Bayesian approach, while still relatively simple, would not be as understandable as our sliding window model.
